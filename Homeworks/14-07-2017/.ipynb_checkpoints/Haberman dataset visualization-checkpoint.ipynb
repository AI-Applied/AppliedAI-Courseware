# Haberman Dataset Analysis


```python
#Imports
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
import seaborn as sb
import matplotlib.pyplot as plt
from pandas.tools.plotting import andrews_curves,parallel_coordinates,radviz
```

## About Haberman Dataset

**Title**: Haberman's Survival Data<br>
**Sources**: (a) Donor: Tjen-Sien Lim (limt@stat.wisc.edu) (b) Date: March 4, 1999<br>
**Past Usage**:<br>
Haberman, S. J. (1976). Generalized Residuals for Log-Linear Models, Proceedings of the 9th International Biometrics Conference, Boston, pp. 104-122.<br>
Landwehr, J. M., Pregibon, D., and Shoemaker, A. C. (1984), Graphical Models for Assessing Logistic Regression Models (with discussion), Journal of the American Statistical Association 79: 61-83.<br>
Lo, W.-D. (1993). Logistic Regression Trees, PhD thesis, Department of Statistics, University of Wisconsin, Madison, WI.
Relevant Information: The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.<br>
Number of Instances: 306<br>
Number of Attributes: 4 (including the class attribute)<br>
**Attribute Information**:<br>
* Age of patient at time of operation (numerical)<br>
* Patient's year of operation (year - 1900, numerical)<br>
* Number of positive axillary nodes detected (numerical)<br>
* Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 year<br>
* Missing Attribute Values: None <br>

```python
#Loading Dataset
data = pd.read_csv('haberman.csv')

data.head()
```

```python
#dataset overview
print("Shape of the data is " +str(data.shape))
print("Headings of the data are " + ','.join('"'+i.upper()+'"' for i in data.columns))
# data['survived'].value_counts()
a=list(data['survived'].value_counts())
#unbalanced
print("Number of people who lived and died within five years are {0} and {1} respectively".format(a[0],a[1]))
```

### Plots

```python
#plotting using pandas methods
data.plot(kind='scatter', x='year', y='axillary')
plt.show()
```

```python
#Using seaborn for plotting
sb.FacetGrid(data, hue="survived", size=5) \
   .map(plt.scatter, "age", "axillary") \
   .add_legend();
plt.show();
```

## Pair Plots

```python
sb.pairplot(data, hue="survived", size=3, diag_kind="kde");
plt.show()
```

```python
#Box Plot
sb.boxplot(x="survived", y="axillary", data=data)
sb.stripplot(x="survived", y="axillary", data=data, jitter=True, edgecolor="gray")

plt.show()
```

```python
#Violin plot
sb.violinplot(x="survived", y="axillary", data=data, size=6)
plt.show()
```

```python
#
sb.FacetGrid(data, hue="survived", size=6) \
   .map(sb.kdeplot, "axillary") \
   .add_legend()
plt.show()
```

```python
sb.pairplot(data, hue="survived", size=3)
plt.show()
```

```python
sb.pairplot(data, hue="survived", size=3, diag_kind="kde")
plt.show()
```

```python
data.boxplot(by='survived',figsize=(12,6))
plt.show()
```

```python
#andrew-curves plotting
andrews_curves(data,'survived')
plt.show()
```

```python
#parallel coordinates plotting
parallel_coordinates(data,'survived')
plt.show()
```

```python
#radviz plotting
radviz(data,'survived')
plt.show()
```

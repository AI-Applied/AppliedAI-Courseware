```python
#Information about the data

print("shape of the data is " + str(data.shape) + "/n")
print("Headers of the data are " + str(list(data.columns)))
# Was unable to perform data["SUBJECTS].value_count()-->why?
#Balanced dataset
print("Number of Data points for each subject \n " + str(Counter(data['SUBJECTS']) ))
```
